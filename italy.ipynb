```python
import pandas as pd
import numpy as np

dta = pd.read_csv(r'C:\Users\jenso\Desktop\Italy2.csv')
dta
```

```python
#renaming the columns
dta.columns = ['year_month', 'Italy CPI', 'Italy IP', 'Italy IP SA', 'US CPI', 'EX']

dta
```

```python
#Italy

#(i)taking the logs of the EX 
dta['EX'] = pd.to_numeric(dta['EX'])
dta['log_EX'] = np.log(dta['EX'])
#(ii) taking the difference in logEX to be diff_log_EX
dta['dif_log_EX'] = dta['log_EX'].diff().fillna(0)

#Getting the inflation rate 
#(iii) log inflation
dta['Italy CPI'] = pd.to_numeric(dta['Italy CPI'])
dta['ITA log CPI'] = np.log(dta['Italy CPI'])
#inflation versus the previous month
dta['monthly ITA inflation'] = dta['ITA log CPI'].diff().fillna(0)



#(iv) log of industrial production
dta['log Italy IP'] = np.log(dta['Italy IP'])
#(v) growth of industrial production
dta['diff_log Italy IP']= dta['log Italy IP'].diff().fillna(0)

pd.options.display.max_rows
dta
                     
```

```python
#US
#(i) log inflation
dta['US_log_CPI'] = np.log(dta['US CPI'])
#(ii) inflation versus the previous month
dta['monthly US inflation'] = dta['US_log_CPI'].diff().fillna(0)
dta
```
